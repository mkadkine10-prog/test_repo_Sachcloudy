```python

from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import *


spark=SparkSession.builder.appName("2ndHighestSalary").getOrCreate()

data = ([1,100],[2,200],[3,300])

schema= (["id","salary"])

df = spark.createDataFrame(data=data,schema=schema)



"""Without Window function:"""

max_salary=df.agg(max('salary')).collect()[0][0]
filtered_df = df.filter(col("salary")!= max_salary)
second_highest_salary=filtered_df.agg(max("salary").alias("second_highest_salary"))
second_highest_salary.show()

"""--------------------------------------------

    using window function :-

    window_spec=Window.orderBy(col("salary").desc())

    ranked_df=df.withColumn("rank",dense_rank().over(window_spec))
    second_highest_salary_df= ranked_df.filter(col("rank")==2).select("id",col("salary").alias("second_highest_salary"))

    second_highest_salary_df.show()

"""


```

```python
window_spec = Window.orderBy(col("salary").desc())
ranked_df = df.withColumn("rank", dense_rank().over(window_spec))
fifth_highest_salary_df = ranked_df.filter(col("rank") == 5).select("id", col("salary").alias("fifth_highest_salary"))
display(fifth_highest_salary_df)
```
